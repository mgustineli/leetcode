```python
"""
NOTE: RUN THIS AND THEN DELETE THIS CELL
"""

import pandas as pd
import numpy as np

names = [
    "Siam Orchid",
    "Golden Basil",
    "Bangkok Flame",
    "Lemongrass Garden",
    "Thai Silk House",
    "Pad Thai Junction",
    "The Mango Leaf",
    "Red Curry Bowl",
    "Thai Elephant Kitchen",
    "Jasmine Spoon",
    "Chiang Mai Table",
    "The Thai Peacock",
    "Green Papaya Grill",
    "Tuk Tuk Thai",
    "Thai Moon Café",
    "Bamboo Ginger",
    "Curry Jasmine",
    "Thai Lantern",
    "Sweet Basil House",
    "Thai Blossom",
    "Royal Thai Spoon",
    "Coconut Grove Thai",
    "Tiger Lily Thai",
    "Street Noodle Bar",
    "Saffron & Lime",
    "Golden Wok Thai",
    "Bangkok Lotus",
    "The Curry Orchid",
    "Thai Harvest",
    "Urban Tuk Tuk",
    "Thai Street Flame",
    "Bangkok Soul Kitchen",
    "Thai Table Collective",
    "Basil & Bird",
    "Thai Tango",
    "Ginger Lemongrass",
    "Sticky Rice House",
    "Thai Khao Kitchen",
    "Midnight Pad Thai",
    "Lotus Breeze",
    "Thai Avenue",
    "Sweet Chilli Thai",
    "Thai Territory",
    "The Golden Mango",
    "Basil Breeze",
    "Thai Craft Kitchen",
    "Lemongrass Street",
    "Chiang Flavor",
    "Lotus & Curry",
    "Thai Serene",
    "Coconut Curry Co.",
    "Mango Orchid",
    "Pad Thai Republic",
    "Siam Street Kitchen",
    "Tamarind Thai",
    "Thai Orchid Café",
    "Bangkok Spice Trail",
    "Lime & Coconut Thai",
    "The Thai Nest",
    "Ginger Bloom",
    "Thai Spoon & Wok",
    "The Spicy Lotus",
    "Chiang Mai Spoon",
    "Thai Garden Lane",
    "Rice Bowl Thai",
    "Mango Root Thai",
    "Basil Street Kitchen",
    "Thai Ember",
    "The Thai Noodle Spot",
    "Lemongrass Bloom",
    "Thai Glow",
    "Urban Thai Garden",
    "Rice Blossom",
    "Thai Harmony",
    "Red Chili Thai",
    "Bangkok Spice Co.",
    "Thai Whisk",
    "The Bamboo Table",
    "Sweet Rice Kitchen",
    "Thai Mingle",
    "Curry Petal",
    "Thai Village Café",
    "The Basil Lamp",
    "Spiced Thai House",
    "Curry Craft Thai",
    "Thai Flame & Wok",
    "Bangkok Rice Bowl",
    "The Thai Cart",
    "Chiang Herb Thai",
    "Jasmine Tiger",
    "Noodle Bloom",
    "Thai Soul Kitchen",
    "Siam Sizzle",
    "Bangkok Thai Yard",
    "Wild Thai Spoon",
    "Thai Bungalow",
    "Basil Noodle Co.",
    "Jasmine Thai Street",
    "Thai Tabletop",
    "The Chili Grove",
    "Lotus Pad Thai",
    "Golden Leaf Thai",
    "Thai Sip & Spoon",
    "Bangkok Harmony",
    "Thai River Flame",
    "Sweet Tamarind",
    "The Thai Petal",
    "Chiang Mai Deli",
    "Thai Lotus Flame",
    "The Wok Nest",
    "Red Basil Kitchen",
    "Thai House of Spice",
    "Mango & Chili",
    "Thai Garden Glow",
    "Thai Flame District",
    "The Curry Lily",
    "Bangkok Tabletop",
    "The Thai Spoon",
    "Ginger Chili Thai",
    "Thai Fire Wok",
    "Bamboo Lemongrass",
    "Thai Garden Kitchen",
    "Thai Breeze House",
    "Spicy Mango Grill",
    "The Golden Wok",
    "The Jasmine Nest",
    "Thai Crave",
    "Thai Rice Market",
    "Bangkok Blossom",
    "Thai Taste Co.",
    "Mango Basil Thai",
    "Thai Street Vibe",
    "Lime Leaf Thai",
    "Thai Spoonful",
    "Thai Bloom Collective",
    "Thai Forest Kitchen",
    "Urban Thai Orchid",
    "Thai Basil District",
    "Thai Spice Route",
    "Lotus Thai Table",
    "Spiced Wok",
    "Jasmine Garden Thai",
    "Bangkok Bamboo",
    "The Thai Bean",
    "Chili Petal Thai",
    "Thai Fireleaf",
    "Thai Basil Café",
    "Golden Noodle House",
    "Mango Thai Co.",
    "Thai Craft Flame",
    "Coconut Curry Café",
    "The Thai Cradle",
    "Thai Urban Bowl",
    "Jasmine Lantern",
    "Thai Curry Hive",
    "Bangkok Wok Co.",
    "Thai Spoon House",
    "Pad Thai Bloom",
    "Thai Orchid Grove",
    "Thai Silk District",
    "Siam Rice House",
    "Thai Herb Spot",
    "Bangkok Flavor Lane",
    "Chili Orchid Thai",
    "The Bamboo Fork",
    "Urban Lotus Thai",
    "Thai Essence",
    "Curry Orchid Grill",
    "Thai Kitchen & Leaf",
    "Thai Bowl & Bloom",
    "Sweet Thai Leaf",
    "Thai Lime Table",
    "Mango Street Thai",
    "Thai Lantern Kitchen",
    "Coconut Thai Fork",
    "Thai Silk Fork",
    "Thai Garden Club",
    "Jasmine Pad Thai",
    "The Curry Grove",
    "Thai Flame Garden",
    "Ginger Leaf Thai",
    "Thai Moonlight",
    "Thai Palate",
    "The Basil Cart",
    "Thai Tandoor",
    "Spicy Thai Trails",
    "Thai Grove Street",
    "Bamboo Thai Craft",
    "Thai Flavor Kitchen",
    "Bangkok Bites",
    "Mango Root Café",
    "Thai Basil Roots",
    "The Curry Table",
    "Chiang Chill Thai",
    "Basil Wok Grill",
    "Thai Crave Street",
    "Golden Thai Nest",
    "Thai Flavor Bloom",
    "Bangkok Basil Leaf",
    "Thai Ember Kitchen",
    "Thai Garden Feast",
    "Urban Jasmine Thai",
    "Thai Mint Bowl",
    "Mango Thai Spoon",
    "Wok Thai Bistro",
    "Lemongrass Heaven",
    "Thai Essence Grill",
    "Curry Thai Spoon",
    "The Thai Grove",
    "Thai Lantern District",
    "Thai Bloom Table",
    "Thai Serenity Café",
    "Rice & Spice Thai",
    "Pad Thai Spoon",
    "Thai Jasmine Lane",
    "Coconut Grove Grill",
    "Thai Fire Pot",
    "Thai Urban Bite",
    "Golden Spoon Thai",
    "Thai Roots Café",
    "Spicy Rice Thai",
    "Jasmine Thai Pantry",
    "Thai Tastes Kitchen",
    "Thai Bamboo Nest",
    "Thai Bowl Flame",
    "The Basil Root",
    "Thai Table Bloom",
    "The Thai Lantern",
    "Ginger Thai Trail",
    "Thai Flavor Nest",
    "Thai Curry Leaf",
    "Bangkok Wok Trail",
    "Thai Garden Street",
    "Mango Petal Thai",
    "The Thai Ember",
    "Thai Blossom Spot",
    "Thai Saffron Bowl",
    "The Basil Spoon",
    "Thai Garden Spoon",
    "Coconut Thai Co.",
    "Thai Bloom Grove",
    "Jasmine Thai Wok",
    "Thai Sunflower",
    "Thai Bowl District",
    "Golden Thai Craft",
    "Pad Thai House",
    "Bamboo Thai Café",
    "The Thai Market",
    "Thai Mingle Spot",
    "Thai Trail Table",
    "Thai Whisper Kitchen",
    "Coconut Thai Bloom",
    "Thai Orchard Lane",
    "Thai Petal Flame",
    "Thai River Table",
    "Thai Harmony Spot",
    "The Noodle Bloom",
    "Thai Basil Tides",
    "Mango Bliss Thai",
    "Thai Table & Root",
    "Bangkok Spoonful",
    "Lemongrass Thai Fork",
    "Thai Heaven Table",
    "The Thai Shed",
    "Thai Craft Wok",
    "Golden Thai Petal",
    "Thai Blossom Lane",
    "Jasmine Thai Craft",
    "Thai Palace Spoon",
    "Thai Bliss House",
    "Thai Curry Bloom",
    "The Thai Lab",
    "Ginger Thai House",
    "Thai Chili Grove",
    "Thai Table Garden",
    "Thai Fusion Grill",
    "Bangkok Roots",
    "Thai Table Hive",
    "Mango & Basil Thai",
    "Thai Sizzle Street",
    "Coconut Thai District",
    "Thai Flavor Lab",
    "Lotus Thai Kitchen",
    "Thai Curry Market",
    "Thai Fusion Nest",
    "Jasmine Thai Spot",
    "Thai Ember Grove",
    "Bamboo Thai Feast",
    "Thai Herb Flame",
    "The Thai Bowl Co.",
    "Thai Blossom Trail",
    "Thai Saffron Spoon",
    "Thai Lemongrass Table",
    "Thai River Grove",
    "The Thai Hatch",
    "Thai Basil Plate",
    "Thai Petal Market",
    "Thai Street Feast",
    "Thai Chili Leaf",
    "Golden Basil Thai",
    "Bangkok Fire Fork",
    "Thai Taste House",
    "Pad Thai Garden",
    "Thai Bungalow Spoon",
    "The Curry Cart",
    "Thai Lime Nest",
    "Thai Mango Table",
    "Coconut Thai Market",
    "Thai Fork District",
    "The Basil Pantry",
    "Thai Harvest Spoon",
    "Thai Cradle Kitchen",
    "Mango Thai District",
    "Thai Lantern Lane",
    "Bangkok Thai Collective",
    "Thai Silk Spoon",
    "Thai Petal Grove",
    "Thai Rice Bloom",
    "Thai Ember Table",
    "Ginger Thai Nest",
    "The Thai Trail",
    "Thai Essence House",
    "Thai Sunset Spoon",
    "Thai Rice Grove",
    "Thai Street Petal",
    "Thai Wok Garden",
    "Thai Mango Blossom",
    "The Thai Door",
    "Bamboo Thai Root",
    "Thai Wok & Spoon",
    "Thai Jasmine Ember",
    "Thai Curry Road",
    "The Noodle Pot",
    "Thai Serene Flame",
    "Pad Thai Porch",
    "Thai Fusion Street",
    "The Thai Fork",
    "Bangkok Thai Kitchen",
    "Thai Bloom Lantern",
    "Thai Garden Fork",
    "Thai Petal Street",
    "Thai Noodle Garden",
    "The Thai Pad",
    "Thai Lime Grove",
    "Jasmine Thai Flame",
    "Thai Mango Lane",
    "Bangkok Rice Pot",
    "Thai Spoon Market",
    "Thai Basil Ember",
    "Thai Orchard Garden",
    "Thai Spice Trail",
    "Thai Curry Fork",
    "Thai Garden Tides",
    "Thai Bloom Spoon",
    "Coconut Thai Table",
    "Thai Noodle Craft",
    "The Thai Flavor",
    "Thai Street Cradle",
    "Thai Leaf District",
    "Thai Sizzle Grove",
    "Thai Fork House",
    "Thai Lemongrass Garden",
    "Thai Bloom & Basil",
    "Thai Harvest Bowl",
    "Thai Curry Haven",
    "The Thai Tabletop",
    "Bangkok Thai Bliss",
    "Thai Bowl Collective",
    "Thai Noodle Spot",
    "Thai Fire Spoon",
    "Thai Lantern Grove",
    "Thai Street Ember",
    "Thai Pad Grove",
    "Thai Mint Spoon",
    "Thai Riverhouse",
    "Thai Wok & Root",
    "Thai Bloom Pantry",
    "Thai Curry Petal",
    "Thai Mango Flame",
    "Bangkok Thai Co.",
    "Thai Garden Cradle",
    "Thai Rice Pot",
    "Thai Bowl Trail",
    "Thai Sunset Bloom",
    "The Thai District",
    "Thai Basil Market",
    "Thai Petal Flame",
    "Thai Orchid Table",
    "Thai Craft Grove",
    "Thai Fork & Leaf",
    "The Thai Collective",
    "Thai Cradle Bowl",
    "Thai Mango Haven",
    "Thai Sizzle Pot",
    "Thai Nest Bloom",
    "Thai Sunset Lane",
    "Thai Curry Spot",
    "Thai Garden Table",
    "Thai Basil Cradle",
    "Thai Mango Co.",
    "The Thai Street",
    "Thai Wok Bloom",
    "Thai Spoon Garden",
    "Thai Lantern Trail",
    "Thai Rice Nest",
    "Thai Curry Pantry",
    "Thai Mango Cradle",
    "Thai Basil Bowl",
    "Thai Ember Haven",
    "Thai Leaf Spot",
    "The Thai Bowl",
    "Thai Garden Nest",
    "Thai Sizzle Fork",
    "Thai Petal Dish",
    "Thai Spoon Cradle",
    "Thai River Pot",
    "Thai Lime Grove",
    "Thai Flame Bowl",
    "Thai Urban Market",
    "Thai Basil Garden",
    "Thai Pad District",
    "Thai Sunset Bowl",
    "Thai Spoon Collective",
    "Thai Herb Lane",
    "Thai Kitchen Grove",
    "Thai Curry Grove",
    "Thai Nest District",
    "Thai Saffron Market",
    "Thai Chili Spot",
    "Thai Rice Blossom",
    "Thai Bloom House",
    "Thai Garden Cart",
    "Thai Curry Table",
    "Thai Basil Bliss",
    "Thai Blossom Flame",
    "Thai Street Bowl",
    "Thai Spoon Trail",
    "Thai Sizzle Dish",
    "Thai Lantern Pot",
    "Thai Petal House",
    "Thai Bowl Haven",
    "Thai Wok Petal",
    "Thai Noodle Grove",
    "Thai Root Pantry",
    "Thai Blossom Table",
    "Thai Essence Grove",
    "Thai Chili Nest",
    "Thai Lime Table",
    "Thai Fire Blossom",
    "Thai Lantern Bliss",
    "Thai Kitchen Bowl",
    "Thai Spoon Nest",
    "Thai Sunset Table",
    "Thai Mango Spot",
    "Thai Flame Nest",
    "Thai Basil Trail",
    "Thai Curry Nest",
    "Thai Bloom Fork",
    "Thai Mango Spoon",
    "Thai Rice Street",
    "Thai Pad Spoon",
    "Thai Garden Co.",
    "Thai Ember Spot",
    "Thai Nest Fork",
    "Thai Wok Table",
    "Thai Rice Trail",
    "Thai Lime Spoon",
    "Thai Lantern Dish",
    "Thai Fire Co.",
    "Thai Root Dish",
    "Thai Garden Bloom",
    "Thai Spoon Fire",
    "Thai Noodle Nest",
    "Thai Petal Trail",
    "Thai Basil Co.",
    "Thai Sunset Nest",
    "Thai Spoon Road",
    "Thai Cradle Spot",
    "Thai Bowl Street",
    "Thai Garden Craft",
    "Thai Lantern Lane",
    "Thai River Co.",
    "Thai Jasmine Street",
    "Thai Spoon Bliss",
    "Thai Market Bowl",
    "Thai Wok Grove",
    "Thai Petal Co.",
    "Thai Fire Garden",
    "Thai Spoon Dish",
    "Thai Chili Grove",
    "Thai Sunset Lane",
    "Thai Leaf Bloom",
    "Thai Ember Bowl",
    "Thai Noodle Spot",
    "Thai Mango Grove",
    "Thai Flame Trail",
    "Thai Curry Leaf",
    "Thai Garden Haven",
    "Thai Spoon Orchard",
    "Thai Lantern Pantry",
]

names = list(set(names))

np.random.seed(42)
n = len(names)

# Numerical features
price = np.random.normal(loc=15, scale=5, size=n).round(2)  # Average meal price
ambience_score = np.clip(np.random.lognormal(mean=2, sigma=0.6, size=n), 1, 5)
service_score = np.clip(np.random.normal(loc=3, scale=1.2, size=n), 1, 5)
food_quality = np.clip(np.random.normal(loc=3.5, scale=0.8, size=n), 1, 5)
aggregate_score = ambience_score + service_score + food_quality

# Add collinear variable
staff_friendliness = service_score + np.random.normal(
    0, 0.3, n
)  # high collinearity with service_score

# Categorical features
locations = ["Downtown", "Suburb", "Mall", "Chinatown", "Beach"]
location = np.random.choice(locations, size=n)

types = ["Sit-down", "Takeout", "Food truck", "Buffet", "Pop-up Kitchen"]
restaurant_type = np.random.choice(types, size=n)

greasy = [True, False]
is_greasy = np.random.choice(greasy, size=n)

# Random noise variable (should not correlate with rating)
num_bathrooms = np.random.randint(1, 4, size=n)

# Construct rating as a function of food_quality, ambience_score, service_score
rating = (
    0.4 * food_quality
    + 0.2 * ambience_score
    + 0.3 * service_score
    + np.random.normal(0, 0.3, n)  # add noise
)
rating = np.clip(np.round(rating), 1, 5)

# Inject some outliers
price[np.random.choice(n, 5)] *= 3  # extreme price
food_quality[np.random.choice(n, 5)] = 1  # bad food
service_score[np.random.choice(n, 5)] = 5  # perfect service

# Inject missing values
for col in ["price", "ambience_score", "location"]:
    ix = np.random.choice(n, size=10, replace=False)
    if col == "location":
        location[ix] = None
    else:
        exec(f"{col}[ix] = np.nan")

# Final DataFrame
df = pd.DataFrame(
    {
        "restaurant_name": names,
        "price": price,
        "ambience_score": ambience_score,
        "service_score": service_score,
        "staff_friendliness": staff_friendliness,
        "food_quality": food_quality,
        "aggregate_score": aggregate_score,
        "location": location,
        "restaurant_type": restaurant_type,
        "num_bathrooms": num_bathrooms,
        "is_greasy": is_greasy,
        "rating": rating.astype(int),
    }
)

# Add duplate rows
duplicates = df.sample(n=20)
df = pd.concat([df, duplicates], ignore_index=True)


# Shuffle rows
df = df.sample(frac=1).reset_index(drop=True)

"""
DATASET CHARACTERISTICS:
Rating is a function of food_quality, ambience_score, service_score

Missing values: Present in price, ambience_score, and location
Outliers: Injected into price, food_quality, and service_score
Collinearity: Between service_score and staff_friendliness
Duplicates: Added via appended sampled rows
Different distributions:
    Normal: price, service_score, food_quality
    Lognormal: ambience_score
    Discrete: num_bathrooms
    Binary: is_greasy
Clipping effects: Scores are bounded in range [1, 5]
"""
```
