````python
# valid palindrome: easy
# https://neetcode.io/problems/is-palindrome?list=neetcode150


def isPalindrome(s: str) -> bool:
    string = [let.lower() for let in s if let.isalnum()]
    return string == string[::-1]


s = "Was it a car or a cat I saw?"
isPalindrome(s)
````

````python
# group anagrams: medium
# https://neetcode.io/problems/anagram-groups?list=neetcode150
from collections import defaultdict


def groupAnagrams(strs: list[str]) -> list[list[str]]:
    # group into sublists
    # sort each word, use as key in hashmap
    group = defaultdict(list)
    for word in strs:
        sort_word = "".join(sorted(word))
        group[sort_word].append(word)
    return list(group.values())


strs = ["act", "pots", "tops", "cat", "stop", "hat"]
output = groupAnagrams(strs)
print(output)
````

````python
# best time to buy and sell stock: easy
# https://neetcode.io/problems/buy-and-sell-crypto?list=neetcode150


def maxProfit(prices: list[int]) -> int:
    min_buy = prices[0]
    maxp = 0
    for price in prices:
        sell = price - min_buy
        maxp = max(maxp, sell)
        min_buy = min(price, min_buy)
    return maxp


prices = [10, 1, 5, 6, 7, 1]
output = maxProfit(prices)
output
````

````python
# remove duplicates: easy
# https://neetcode.io/problems/remove-duplicates-from-sorted-array?list=neetcode150


def removeDuplicates(nums: list[int]) -> int:
    left = 1
    for right in range(1, len(nums)):
        if nums[right] != nums[right - 1]:
            nums[left] = nums[right]
            left += 1
    return left


nums = [2, 10, 10, 30, 30, 30]
index = removeDuplicates(nums)
nums[:index]
````

````python
# valid anagram: easy
# https://neetcode.io/problems/is-anagram?list=neetcode150


def isAnagram(s: str, t: str) -> bool:
    if len(s) != len(t):
        return False

    def count_letters(arr):
        hashmap = {}
        for let in arr:
            hashmap[let] = 1 + hashmap.get(let, 0)
        return hashmap

    hash_s = count_letters(s)
    hash_t = count_letters(t)

    return hash_s == hash_t


s = "racecar"
t = "carrace"
output = isAnagram(s, t)
output
````

````python
# longest consecutive sequence
# https://neetcode.io/problems/longest-consecutive-sequence?list=neetcode150


def longestConsecutive(nums: list[int]) -> int:
    numSet = set(nums)
    longest = 0

    for num in numSet:
        if (num - 1) not in numSet:
            length = 1
            while (num + length) in numSet:
                length += 1
            longest = max(length, longest)
    return longest


nums = [2, 20, 4, 10, 3, 4, 5]
output = longestConsecutive(nums)
output
````

````python
# container with most water: medium
# https://neetcode.io/problems/max-water-container?list=neetcode150
def maxArea(heights: list[int]) -> int:
    max_water = 0
    left, right = 0, len(heights) - 1
    while left < right:
        area = (right - left) * min(heights[left], heights[right])
        max_water = max(max_water, area)
        if heights[left] < heights[right]:
            left += 1
        else:
            right -= 1
    return max_water


# heights = [2, 2, 2]
heights = [1, 7, 2, 5, 4, 7, 3, 6]
output = maxArea(heights)
output
````

````python
# top k frequent elements: medium
# https://neetcode.io/problems/top-k-elements-in-list?list=neetcode150


def topKFrequent(nums: list[int], k: int) -> list[int]:
    hashmap = {}
    for num in nums:
        hashmap[num] = 1 + hashmap.get(num, 0)

    arr = sorted(hashmap.items(), key=lambda item: item[1], reverse=True)
    output = [k for k, v in arr[:k]]
    return output


nums = [1, 2, 2, 3, 3, 3]
output = topKFrequent(nums, k=2)
output
````

````python
# Zoox interview question: recenter trajections
import torch

trajections = torch.tensor(
    [
        [[1, 2], [2, 2], [3, 2]],
        [[1, 2], [2, 2], [3, 2]],
    ]
)


def recenter(trajections):
    new_trajections = trajections.clone()
    new_trajections -= torch.unsqueeze(trajections[0, 0], dim=0)
    return new_trajections


print(trajections.shape)
new_trajections = recenter(trajections)
print(new_trajections)
print(new_trajections.shape)
````

Given a binary tree where each node contains an integer value, determine if the tree has any **root-to-leaf** path such that the sum of the values along the path equals a given `targetSum`.

A _root-to-leaf_ path is defined as a path starting from the root node and ending at a leaf node (a node with no children).

```text
      5
     / \
    4   8
   /   / \
 11  13  4
/  \       \
7    2       1
```

```python
targetSum = 22
```

**Expected output:**
```python
True
```

**Explanation:** The path 5 → 4 → 11 → 2 sums up to 22.

````python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def search_tree(root, target_sum, total_sum):
    if not root:
        return False

    total_sum += root.val

    if not root.left and not root.right:
        return total_sum == target_sum

    return search_tree(root.left, target_sum, total_sum) or search_tree(
        root.right, target_sum, total_sum
    )


tree = TreeNode(5)
tree.left = TreeNode(4)
tree.right = TreeNode(8)
tree.left.left = TreeNode(11)
tree.left.left.left = TreeNode(7)
tree.left.left.right = TreeNode(2)
tree.right.left = TreeNode(13)
tree.right.right = TreeNode(4)
tree.right.right.right = TreeNode(1)

print(search_tree(tree, 22, 0))  # Output: True
````
